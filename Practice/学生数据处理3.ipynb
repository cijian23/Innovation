```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_name):
    df = pd.read_csv("D:/Innovation/DataSet/" + file_name ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
df = load_data('lai_temp.csv')
df
```

```python
df['商户名称'].value_counts()
```

```python
name = df['商户名称']
name = name.to_list()
```

```python
name
```

```python
name = [x.strip() for x in name]
name
```

```python
df['商户名称'] = name
df
```

```python
df['商户名称'].value_counts()
```

```python
name_index = {}

for i, word in enumerate(df['商户名称'].value_counts().index):
    name_index[word] = i
    
name_index
```

```python
name_index['美食林肠粉']
```

```python
name = pd.DataFrame(list(name_index.items()), columns=['word', 'index'])
name.to_csv('D:/Innovation/DataSet/商户名称.csv', encoding='gbk', index=False)
```

```python
name = []

for n in df['商户名称']:
    for key, value in name_index.items():
        if n == key:
            name.append(value)
            
name
```

```python
df['商户名称(编号)'] = name
df
```

```python
X = pd.get_dummies(df['商户名称'])
X
```

```python
column = ['商户名称','星期','是否为工作日','时段','天气好/坏']
# 一个一个的进行处理，并连接到数据集的后面
for i in column:
    x =pd.get_dummies(df[i],prefix= i)
    df=pd.concat([df,x],axis=1)
# 显示转换后的数据
df.head()
```

```python
df.to_csv('D:/Innovation/DataSet/one-hot.csv', encoding='gbk', index=False)
```
