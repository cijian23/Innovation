```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_name):
    df = pd.read_csv("D:/大创/temp/" + file_name ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
df = load_data('赖（整时处理2）.csv')
df
```

```python
from chinese_calendar import is_holiday
import datetime

holiday = []

for date in df['交易日期'] :
    date = datetime.datetime.strptime(date, "%Y/%m/%d").date()
    if is_holiday(date):
        holiday.append(1)
    else :
        holiday.append(0)

df['节假日'] = holiday
df[:20]
```

```python
climate = load_data('climate.csv')
climate
```

```python
plt.style.use("fivethirtyeight")
#设置子图
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
fig.autofmt_xdate(rotation=45)
ax1.plot(climate["气温"],linewidth= 0.2)
ax1.set_xlabel(''); ax1.set_ylabel("Temperature"); ax1.set_title("Temp")
ax2.plot(climate["相对湿度"],linewidth= 0.2)
ax2.set_xlabel(''); ax2.set_ylabel("Humidity"); ax2.set_title("Humidity")
ax3.plot(climate["气象站大气压"],linewidth= 0.2)
ax3.set_xlabel("Date"); ax3.set_ylabel("Ground Press"); ax3.set_title("Ground Press")
ax4.plot(climate["海平面大气压"],linewidth= 0.2)
ax4.set_xlabel("Date"); ax4.set_ylabel("Sea Press"); ax4.set_title("Sea Press")
plt.tight_layout(pad=2)
plt.show()
```

```python
climate['时间'] = climate['时间'].str.zfill(5)
climate[160:190]
```

```python
df['日期小时'] = df['交易日期'] + ' ' + df['交易时间(整时)'].str[:2]
climate['日期小时'] = climate['date'] + ' ' + climate['时间'].str[:2]

merged_df = pd.merge(df, climate, on='日期小时')
merged_df
```

```python
merged_df = merged_df.drop(['学工号','姓名','日期小时','date','时间','重要天象','交易时间(整时)'],axis=1)
merged_df
```

```python
merged_df.to_csv('D:/大创/temp/赖（合并天气3）.csv',sep= ',', encoding='gbk',index=False)
```
