```python
def load_data(file_name):
    df = pd.read_csv("D:/Innovation/DataSet/" + file_name ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
workday = []

for date in df['交易日期'] :
    date = datetime.datetime.strptime(date, "%Y/%m/%d").date()
    if is_workday(date):
        workday.append(1)
    else :
        workday.append(0)

workday[0]
```

```python
df['是否为工作日'] = workday
```

```python
def join_feature(feature):
    result = []
    for date_i in df["交易日期"]:
        for date_j in clm['date']:
            if date_i == date_j:
                temp = clm[clm['date'] == date_i]
                result.append(temp[feature].values[0])
    return result
```

```python
temperature = join_feature("气温")
print(temperature[0])
```

```python
df['气温'] = temperature
df
```

```python
surface_p = join_feature('气象站大气压')
ocean_p = join_feature('海平面大气压')
humidity = join_feature('相对湿度')
wind = join_feature('平均风速')
df['气象站大气压'] = surface_p
df['海平面大气压'] = ocean_p
df['相对湿度'] = humidity
df['平均风速'] = wind
df
```
