```python
import pandas as pd
import os
```

```python
def load_data(file_path):
    df = pd.read_csv(file_path ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
file_path = '../temp/final'

#生成路径
lai_path = os.path.join(file_path, 'lai', 'final.csv')
li_path = os.path.join(file_path, 'li', 'final.csv')
chen_path = os.path.join(file_path, 'chen', 'final.csv')
lu_path = os.path.join(file_path, 'lu', 'final.csv')
zheng_path = os.path.join(file_path, 'zheng', 'final.csv')

lai = load_data(lai_path)
li = load_data(li_path)
lu = load_data(lu_path)
chen = load_data(chen_path)
zheng = load_data(zheng_path)
```

```python
df = pd.concat([lai, li, lu, chen, zheng], copy= True)
df
```

```python
df['商户编号'] = pd.factorize(df['商户名称'])[0]

# 创建映射关系DataFrame
mapping_df = pd.DataFrame({
    '商户名称': df['商户名称'].unique(),
    '商户编号': df['商户编号'].unique()
})


# 将映射关系保存为CSV文件
mapping_df.to_csv('map.csv', sep= ',', encoding='gbk', index=False)
```

```python
column_mapping = {
    '交易额': 'amount',
    '商户名称': 'merchant_name',
    '交易日期': 'date',
    '交易时间': 'time',
    '星期': 'weekday',
    '节假日': 'holiday',
    '气温': 'temperature',
    '气象站大气压': 'station_pressure',
    '海平面大气压': 'sea_level_pressure',
    '相对湿度': 'relative_humidity',
    '平均风速': 'windspeed',
    '特殊天象': 'special_weather',
    '时间戳': 'timestamp',
    '饭点课程_早': 'meal_course_breakfast',
    '饭点课程_午': 'meal_course_lunch',
    '饭点课程_晚': 'meal_course_dinner',
    '商户编号': 'merchant_code'
}

# 使用rename()方法重命名DataFrame的列
df.rename(columns=column_mapping, inplace=True)
```

```python
lai = df[:471]
li = df[471:758]
lu = df[758:1088]
chen = df[1088:1383]
zheng = df[1383:]
```

```python
def save_data(data, id):
    save_path = os.path.join(file_path, id, 'final(english).csv')
    data.to_csv(save_path,sep= ',', encoding='gbk',index=False)

save_data(lai, 'lai')
save_data(li, 'li')
save_data(lu, 'lu')
save_data(chen, 'chen')
save_data(zheng, 'zheng')
```
