```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_name):
    df = pd.read_csv("D:/大创/temp/" + file_name ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
df = load_data('赖（去异常1）.csv')
df
```

```python
df['交易时间'] = df['交易日期'] + ' '+ df['交易时间']
df['交易时间'] = pd.to_datetime(df['交易时间'], format='%Y/%m/%d %H:%M')
df
```

```python
#创建一个空的DataFrame或列表来存储处理后的数据。
processed_data = pd.DataFrame(columns=df.columns)

prev_row = None

for index, row in df.iterrows():
    if prev_row is None:
        prev_row = row
    else:
        time_diff = prev_row['交易时间'] - row['交易时间']
        print(time_diff)
        if (time_diff <= pd.Timedelta(minutes=30)) and (row['商户名称'] == prev_row['商户名称']):
            prev_row['交易额'] += row['交易额']
        else:
            processed_data = pd.concat([processed_data, prev_row.to_frame().T], ignore_index=True)
            prev_row = row

# 处理最后一行数据
processed_data = pd.concat([processed_data, prev_row.to_frame().T], ignore_index=True)
```

```python
# 将Series转换为datetime类型
datetime_series = pd.to_datetime(processed_data['交易时间'])

# 提取小时和分钟部分
hour_minute_series = datetime_series.dt.strftime('%H:%M')
processed_data['交易时间'] = hour_minute_series
processed_data
```

```python
from datetime import datetime, timedelta

# 假设您的交易时间数据保存在一个名为"trade_times"的列表中
trade_times = processed_data['交易时间']

processed_times = []

for time_str in trade_times:
    # 将时间字符串转换为datetime对象
    dt = datetime.strptime(time_str, "%H:%M")
    # 如果分钟大于等于30，则小时进一位，并将分钟置为0
    if dt.minute >= 30:
        dt = dt.replace(hour=dt.hour + 1, minute=0)
    else:
        dt = dt.replace(minute=0)
    # 将处理后的时间添加到列表中
    processed_times.append(dt.strftime("%H:%M"))

# 输出处理后的时间列表
print(processed_times)
```

```python
processed_data['交易时间(整时)'] = processed_times
processed_data
```

```python
processed_data.to_csv('D:/大创/temp/赖（整时处理2）.csv',sep= ',', encoding='gbk',index=False)
```
