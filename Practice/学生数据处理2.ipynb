```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_name):
    df = pd.read_csv("D:/Innovation/DataSet/" + file_name ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
df = load_data('混合数据.csv')
df
```

```python
week = []

for w in df['星期']:
    if w == '星期一':
        week.append(1)
    if w == '星期二':
        week.append(2)
    if w == '星期三':
        week.append(3)
    if w == '星期四':
        week.append(4)
    if w == '星期五':
        week.append(5)
    if w == '星期六':
        week.append(6)
    if w == '星期日':
        week.append(7)
        
week
```

```python
df['星期'] = week
df
```

```python
time = df['交易时间']
```

```python
import datetime
```

```python
int(datetime.datetime.strptime(time[2], "%H:%M").strftime("%H"))
```

```python
period = []

for t in time:
    T = int(datetime.datetime.strptime(t, "%H:%M").strftime("%H"))
    if  T >= 6 and T < 10:   #morning
        period.append(0)
    if T >= 10 and T < 13:   #noon
        period.append(1)
    if T >= 13 and T < 17:   #afternoon
        period.append(2)
    if T >= 17 and T < 19:   #evening
        period.append(3)
    if T >= 19 or T < 6:    #night
        period.append(4)
        
period
```

```python
df['时段'] = period
df
```

```python
df.to_csv('D:/Innovation/DataSet/混合数据.csv',sep= ',', encoding='gbk')
```
