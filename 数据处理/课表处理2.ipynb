```python
import pandas as pd
import icalendar
import re
import os
from datetime import datetime
```

```python
# 指定文件路径
file_path = '../temp/final'
id = 'zheng'

#生成课表路径
spring_path = os.path.join(file_path, id, '2023_spring.csv')
autumn_path = os.path.join(file_path, id, '2022_autumn.csv')
summer_path = os.path.join(file_path, id, '2022_summer.csv')

#生成data路径
data_path = os.path.join(file_path, id, 'data.csv')
```

```python
def load_data(file_path):
    df = pd.read_csv(file_path ,encoding='gbk')
    columns = df.columns
    df.fillna(df.mean(numeric_only=True) ,inplace=True)
    return df
```

```python
data = load_data(data_path)
spring = load_data(spring_path)
summer = load_data(summer_path)
autumn = load_data(autumn_path)
```

```python
def week2num(df):
    week = []
    for row in df:
        if row == 'Monday':
            week.append(1)
        elif row == 'Tuesday':
            week.append(2)
        elif row == 'Wednesday':
            week.append(3)
        elif row == 'Thursday':
            week.append(4)
        elif row == 'Friday':
            week.append(5)
        elif row == 'Saturday':
            week.append(6)
        elif row == 'Sunday':
            week.append(7)
    return week
```

```python
spring_week = week2num(spring['Weekday'])
autumn_week = week2num(autumn['Weekday'])
summer_week = week2num(summer['Weekday'])

spring['Weekday'] = spring_week
autumn['Weekday'] = autumn_week
summer['Weekday'] = summer_week
```

```python
data['饭点课程_早'] = 0
data['饭点课程_午'] = 0
data['饭点课程_晚'] = 0
data
```

```python
class_list = pd.concat([spring, autumn], copy=True)
class_list = pd.concat([class_list, summer], copy=True)
class_list[:60]
```

```python
# 将日期字符串转换为日期格式
data['交易日期'] = pd.to_datetime(data['交易日期'], format="%Y/%m/%d")
class_list['Start Date'] = pd.to_datetime(class_list['Start Date'], format="%Y-%m-%d")
class_list['End Date'] = pd.to_datetime(class_list['End Date'], format="%Y-%m-%d")

# 使用条件索引进行高效筛选和修改
for _, row_c in class_list.iterrows():
    mask = (data['交易日期'].between(row_c['Start Date'], row_c['End Date'])) & (data['星期'] == row_c['Weekday'])
    if row_c['Start Time'] == '08:00':
        data.loc[mask, '饭点课程_早'] = 1
    if row_c['End Time'] == '11:45':
        data.loc[mask, '饭点课程_午'] = 1
    if row_c['End Time'] == '17:30':
        data.loc[mask, '饭点课程_晚'] = 1
```
